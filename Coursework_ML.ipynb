```python
# Загружаем библиотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing

from sklearn.model_selection import train_test_split
```

```python
# Загружаем данные
df = pd.read_csv('telecom_users.csv')
```

```python
# Осматриваем датафрейм на предмет формата предоставленных данных,
# лишних столбцов и пропущенных значений
df.info()
```

```python
# Удалим лишние столбцы и сменим тип данных столбца 'SeniorCitizen' на 'object'
drop_elements = 'Unnamed: 0'
df = df.drop(drop_elements, axis = 1)
df = df.astype({'SeniorCitizen':'object'})
```

```python
# Ознакомимся с уникальными значениями признаков
for col in df.columns:
    print('\n', col, '\ntype:',df[col].dtype)
    print(df[col].unique())
```

Столбец 'TotalCharges' содержит данные типа float, однако они определены как object

```python
# Переведём тип значений столбца в float
# пустые или некоректные значения заменим на NaN
# и выведем на экран
count = 0
for i in range(len(df['TotalCharges'])):
    value = df.at[i, 'TotalCharges']
    try:
        df.at[i,'TotalCharges'] = float(value)
    except Exception:
        df.at[i,'TotalCharges'] = np.NaN
        print('\n№', i)
        print(df.iloc[356,:])
        count += 1
print('\n\nИз 5986 значений', count, 'некорректных')
print('\n\n')
```

Как видим некорректных значений мало и они связаны с одним и тем же клиентом. Можем строки их содержащие, удалить из набора данных.

```python
df = df.dropna()
```

Поменяем тип данных столбца 'TotalCharges' с object на float

```python
df = df.astype({'TotalCharges':'float'})
df.info()
```

```python
# также удалим ненужный уже ID
df = df.drop('customerID', axis = 1)
```

```python
# Создадим кодировщик и установим соответствие
# между метками и числами
data = pd.DataFrame(columns=df.columns)
df_labels = []

for col in df.columns:
    if df[col].dtype == 'object':
        print('\nСоздётся кодировщик для: ', col)
        encoder = preprocessing.LabelEncoder()
        print('Кодировщик заполняется  метками... ')
        encoder.fit(df[col].unique())
        
        for i, itern in enumerate(encoder.classes_):
            print(itern, '-->', i)
        print('Заполнение фрейм закодированными метками...')
        data[col] = encoder.transform(df[col])
        print('Сохранение кодировщика и названия столбца')
        df_labels.append([encoder, col])
    else:
        data[col] = df[col].values
```

# 3. Построение моделей для прогнозирования оттока

```python
### YOUR CODE HERE ###
```

```python
# Разобьём данные на входные и целевые
X = data.drop('Churn', axis = 1)
y  = data['Churn']
x_data, x_valid, y_data, y_valid = train_test_split(X, y, test_size=0.1, random_state=17, stratify=y)
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=17, stratify=y_data)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                     StratifiedKFold)

from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
```

Для построения моделей используем следующие классификаторы:
* Дерево решений - DecisionTreeClassifier()
* k-ближайших соседей - KNeighborsClassifier()
* Метод опорных векторов - SVC()
* Логистическая регрессия - LogisticRegression()
* Случайный лес - RandomForestClassifier()
* Градиентный бустинг - GradientBoostingClassifier()

Для построения метамодели используем градиентный бустинг:
* XGBClassifier() библиотеки xgboost

Для оценки точности предсказаний наших моделей будем использовать критерий ROC_AUC, так как он более устойчив к несбалансированным классам.

```python
dtc_tree = DecisionTreeClassifier(random_state=17, class_weight ='balanced')
knn = KNeighborsClassifier()
svc = SVC(gamma='auto', random_state=17, class_weight ='balanced')
lr = LogisticRegression(random_state=17, max_iter=1000, class_weight ='balanced')
rfc = RandomForestClassifier(random_state=17, class_weight ='balanced')
gbc = GradientBoostingClassifier(random_state=17)


skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=17)

xgb = XGBClassifier(random_state=17, verbosity=0, use_label_encoder=False)

dtc_params = {'criterion':['gini', 'entropy'], 
              'max_depth':range(1,5)}

knn_params = {'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
              'weights':['uniform', 'distance'],
              'n_neighbors':range(1, 10)}
        
svc_params = {'kernel': ['linear', 'rbf'],
              'C': np.arange(0.1, 1, 0.2)}

lr_params = {'C': np.arange(0.5, 1, 0.1)}

rfc_params = {'n_estimators': range(10, 100, 10),
              'min_samples_leaf': range(1, 5)}

gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1),
              'loss':['deviance', 'exponential']}


xgb_params = {'n_estimators': range(10, 100, 5),
              'eta': np.arange(0.1, 1., .1),
              'min_child_weight': range(1, 10, 1),
              'subsample': np.arange(0.1, 1., 0.2)}

classifiers = {'dtc_tree':dtc_tree, 'knn':knn, 'svc':svc, 'lr':lr, 'gbc':gbc, 'rfc':rfc}
params = {'dtc_tree':dtc_params, 'knn':knn_params, 'svc':svc_params, 
          'lr':lr_params, 'gbc':gbc_params, 'rfc':rfc_params,}
```

```python
gs_models = []
score = []
for clf in classifiers:
    print('\nПоиск оптимальных параметров для:', classifiers[clf])
    gs_model = GridSearchCV(classifiers[clf], params[clf], cv=skf, scoring='roc_auc').fit(x_train, y_train)
    gs_models.append(gs_model)
    y_pred = gs_model.predict(x_test)
    conf_mtrx = confusion_matrix(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    print('best_params:', gs_model.best_params_)
    print('Матрица ошибок:\n', conf_mtrx)
    print(f'roc_auc_score: {roc_auc}')
    score.append(roc_auc)
```

Как видим, моделям тяжело идентифицировать класс '1' и в случаях, когда они делают это успешно, то у них ухудшается способность распознавать класс '0'.

Попробуем построить метамодель, возможно она покажет лучшие результаты

```python
meta_mtrx = np.empty((x_test.shape[0], len(classifiers))) 
for n, gs_model in enumerate(gs_models):
    y_pred = gs_model.predict(x_test)
    meta_mtrx[:, n] = y_pred
```

```python
print('Поиск оптимальных параметров для метаклассификатора XGBClassifier')
meta = GridSearchCV(xgb, xgb_params, cv=skf, scoring='roc_auc').fit(meta_mtrx, y_test)
print('Лучшие параметры:', meta.best_params_)
```

```python
meta_mtrx_test = np.empty((x_valid.shape[0], len(classifiers))) 
for n, clf in enumerate(gs_models):
    meta_mtrx_test[:, n] = clf.predict(x_valid)
meta_predict = meta.predict(meta_mtrx_test)


roc_auc = roc_auc_score(y_valid, meta_predict)
print(f'roc_auc_score: {roc_auc}')
conf_mtrx = confusion_matrix(y_valid, meta_predict)
print('Матрица ошибок:\n', conf_mtrx)
```

К сожалению, метакласификатор не смог улучшить ситуацию с распознованием класса '1'

```python
roc_auc_max = [[0,0],[0,0],[0,0]]

for n, sc in enumerate(score):
    if sc>roc_auc_max[0][1]:
        roc_auc_max[2][1]=roc_auc_max[1][1]
        roc_auc_max[2][0]=roc_auc_max[1][0]
        roc_auc_max[1][1]=roc_auc_max[0][1]
        roc_auc_max[1][0]=roc_auc_max[0][0]
        
        roc_auc_max[0][1]=sc
        roc_auc_max[0][0]=n

    elif sc>roc_auc_max[1][1]:
        roc_auc_max[2][1]=roc_auc_max[1][1]
        roc_auc_max[2][0]=roc_auc_max[1][0]        
        
        roc_auc_max[1][1]=sc
        roc_auc_max[1][0]=n  
        
        
    elif sc>roc_auc_max[2][1]:
        roc_auc_max[2][1]=sc
        roc_auc_max[2][0]=n    
   
    
print(f'Лучшие показатели:')    
for n , ram in roc_auc_max:
    print(f'\nКлассификатор:', gs_models[n].estimator)
    print(n, 'roc_auc_score:', ram)

print(f'\nС использование метаклассификатора XGBClassifier:')
print(f'roc_auc_score: {roc_auc}')
```
